```python
import pandas as pd
import datetime as dt
import time
```

```python
path = "/home/ubuntu/classify_records.csv"
```

```python
def str_to_timedelta(string):
    t = dt.datetime.strptime(string, "%H:%M:%S")
    return dt.timedelta(hours=t.hour, minutes=t.minute, seconds=t.second)
```

```python
df = pd.read_csv(path, sep='\t', skiprows=None, parse_dates=[2, 3, 4], date_parser=str_to_timedelta)
```

```python
df
```

```python
df.HASHES_SIZE = pd.to_numeric(df.HASHES_SIZE.str.replace("GB", ""))
```

```python
for col in ["TOTAL", "CLASSIFY", "BINNING"]:
    df[col] = df[col].dt.total_seconds()
```
